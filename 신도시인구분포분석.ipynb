#### 위례신도시 인구 데이터 중 0 ~ 100세 이상까지의 인구수 데이터만 추출하여 저장

- 위 데이터를 살펴보면 row[0]에는 지역명, row[1], row[2]에는 해당 지역의 총인구수가 나오고
- 그 뒤로는 0~100세 이상까지의 인구수가 순서대로 저장되어 있음
- 따라서 이 리스트의 3번 인덱스부터 끝까지 데이터를 반복해서 추출해서 저장
- 이때, 위례신도시는 송파구, 성남시, 하남시로 나뉘어 있으므로 각각 나눠서 저장

```python
import csv
f = open('age.csv')
data = csv.reader(f)
songpa = []     # 빈 리스트 만들기
seongnam = []
hanam = []

for row in data:
    if '송파구 위례' in row[0]:     # 송파구 위례가 포함된 행정구역 찾기
        for i in row[3:]:     # 0세부터 100세이상까지 해당 연령의 인구수를 리스트에 순서대로 저장
            songpa.append(int(i.replace(',','')))     # 문자로 저장되어 있는 데이터를 정수로 변경
    elif '성남시 수정구 위례' in row[0]:
        for i in row[3:]:
            seongnam.append(int(i.replace(',','')))
    elif '하남시 위례' in row[0]:
        for i in row[3:]:
            hanam.append(int(i.replace(',','')))
print(songpa)
print(seongnam)
print(hanam)
```

```python
# 송파위례, 성남위례, 하남위례 합친 위례 전체 연령별 인구
wirye = [] 
for i in range(len(songpa)):
    wirye.append(songpa[i] + seongnam[i] + hanam[i])
wirye
```

#### 성별 구분되어 있는 데이터를 다시 다운받아 분석해보기

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
next(data)
```

- data를 살펴보면 남성 데이터 먼저, 그 뒤로 여성데이터가 나옴
- 따라서 0세 ~ 100세 이상 인구수는 남성 데이터의 경우 인덱스가 3에서 시작하고, 여성 데이터의 경우 인덱스가 -1에서 시작해서 추출

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
songpa_m = []
songpa_f = []
seongnam_m = []
seongnam_f = []
hanam_m = []
hanam_f = []

for row in data:
    if '송파구 위례' in row[0]:
        for i in range(0, 101):
            songpa_m.append(-int(row[i + 3]))     # 남여 데이터 막대그래프 그릴 때 구분을 위해 음수로 변경
            songpa_f.append(int(row[-(i + 1)]))
    elif '성남시 수정구 위례' in row[0]:
        for i in range(0, 101):
            seongnam_m.append(-int(row[i + 3]))
            seongnam_f.append(int(row[-(i + 1)]))
    elif '하남시 위례' in row[0]:
        for i in range(0, 101):
            hanam_m.append(-int(row[i + 3]))
            hanam_f.append(int(row[-(i + 1)]))
songpa_f.reverse()     # 리스트의 값을 역순으로 재배열
seongnam_f.reverse()
hanam_f.reverse()
```

```python
# 송파위례, 성남위례, 하남위례 합친 위례 전체 연령별, 성별 인구
wirye_m = []
wirye_f = []
for i in range(len(songpa_m)):
    wirye_m.append(songpa_m[i] + seongnam_m[i] + hanam_m[i])
for i in range(len(songpa_f)):
    wirye_f.append(songpa_f[i] + seongnam_f[i] + hanam_f[i])
print(wirye_m)
print(wirye_f)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), songpa_m, label = '송파_남')
plt.barh(range(101), songpa_f, label = '송파_여')
plt.barh(range(101), seongnam_m, label = '성남_남')
plt.barh(range(101), seongnam_f, label = '성남_여')
plt.barh(range(101), hanam_m, label = '하남_남')
plt.barh(range(101), hanam_f, label = '하남_여')
plt.title('위례신도시 지역의 남녀 성별 인구 분포')
plt.legend()
plt.show()
```

- 위례신도시의 경우 남녀 비율은 비슷한 형태를 띔

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), wirye_m, label = '위례_남')
plt.barh(range(101), wirye_f, label = '위례_여')
plt.title('위례신도시 지역 전체의 남녀 성별 인구 분포')
plt.legend()
plt.show()
```

#### 성별인구 비율을 파이차트(원그래포)로 표현

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
size = []
for row in data:
    if '송파구 위례' in row[0]:
        m = 0
        f = 0
        for i in range(101):
            m += int(row[i + 3])     # 3~103번까지 데이터 더하기
            f += int(row[i + 106])     # 106~206번까지 데이터 더하기
    elif '성남시 수정구 위례' in row[0]:
        for i in range(101):
            m += int(row[i + 3])
            f += int(row[i + 106])
    elif '하남시 위례' in row[0]:
        for i in range(101):
            m += int(row[i + 3])
            f += int(row[i + 106])
size.append(m)     # 남성 합계 데이터 리스트에 추가
size.append(f)     # 여성 합계 데이터 리스트에 추가
print(size)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')     # 한글폰트
color = ['crimson', 'darkcyan']     # 색 지정
plt.axis('equal')
plt.pie(size, labels = ['남', '여'], autopct = '%.1f%%', colors = color, startangle = 90)
# autopct: 비율표시. %.1f%%는 소수점 아래 첫번째 소수점까지 표현. explode: 돌출 효과, startangle: 시작 각도 지정
plt.title('위례신도시의 남녀 성별 비율')
plt.legend()     # 범례표시
plt.show()
```

#### 연령별 성별 데이터를 꺾은선 그래프로 표현

```python
import csv
f = open('gender.csv')
data = csv.reader(f)
songpa_m = []
songpa_f = []
seongnam_m = []
seongnam_f = []
hanam_m = []
hanam_f = []

for row in data:
    if '송파구 위례' in row[0]:
        for i in range(3, 104):
            songpa_m.append(int(row[i]))     # 남성 데이터 저장
            songpa_f.append(int(row[i + 103]))     # 여성 데이터 저장
    elif '성남시 수정구 위례' in row[0]:
        for i in range(3, 104):
            seongnam_m.append(int(row[i]))
            seongnam_f.append(int(row[i + 103]))
    elif '하남시 위례' in row[0]:
        for i in range(3, 104):
            hanam_m.append(int(row[i]))
            hanam_f.append(int(row[i + 103]))
```

```python
import matplotlib.pyplot as plt
plt.plot(songpa_m, label = '송파_남')
plt.plot(songpa_f, label = '송파_여')
plt.legend()
plt.show()
plt.plot(seongnam_m, label = '성남_남')
plt.plot(seongnam_f, label = '성남_여')
plt.legend()
plt.show()
plt.plot(hanam_m, label = '하남_남')
plt.plot(hanam_f, label = '하남_여')
plt.legend()
plt.show()
```

```python
# 송파위례, 성남위례, 하남위례 합친 위례 전체 연령별, 성별 인구
wirye_m = []
wirye_f = []
for i in range(len(songpa_m)):
    wirye_m.append(songpa_m[i] + seongnam_m[i] + hanam_m[i])
for i in range(len(songpa_f)):
    wirye_f.append(songpa_f[i] + seongnam_f[i] + hanam_f[i])
print(wirye_m)
print(wirye_f)
```

```python
import matplotlib.pyplot as plt
plt.plot(wirye_m, label = 'Male')
plt.plot(wirye_f, label = 'Female')
plt.legend()
plt.show()
```

#### 여성인구에서 남성인구를 뺀 값을 막대그래프로 표현

```python
result = []
for i in range(len(wirye_f)):
    result.append(wirye_f[i] - wirye_m[i])
print(result)
```

```python
plt.bar(range(101), result)
plt.show()
```

- 위례신도시에는 대부분의 나이대에서 여성인구가 남성인구보다 많음
- 20대이하와 40대는 주로 남성이 여성보다 많고, 20대 후반 ~ 40대 초반, 50대 이상의 경우에는 여성이 많음

#### 연령대별 성별 비율을 산점도로 표현

```python
plt.scatter(wirye_m, wirye_f, c = range(101), alpha = 0.5, cmap = 'jet')     # 컬러맵 적용
plt.colorbar()
plt.plot(range(max(wirye_m)), range(max(wirye_m)), 'g')     # 추세선 추가
plt.style.use('ggplot')
plt.show()
```

- x축이 남성인구수, y축이 여성인구수
- 위에서 살펴본 것처럼 대부분의 연령대에서 여성인구수가 많음

```python
import math

size = []
for i in range(len(wirye_m)):
    size.append(math.sqrt(wirye_m[i] + wirye_f[i]))     # 연령별 남성과 여성 인구수 합친값을 제곱근함수를 이용해 점의 크기 조절
print(size)
```

```python
plt.style.use('ggplot')
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize = (10, 5), dpi = 300)
plt.title('위례신도시의 성별 인구 그래프')
plt.scatter(wirye_m, wirye_f, s = size, c = range(101), alpha = 0.5, cmap = 'jet')
plt.colorbar()
plt.plot(range(max(wirye_m)), range(max(wirye_m)), 'g')
plt.xlabel('남성 인구수')
plt.ylabel('여성 인구수')
plt.show()
```

- 위에서 다른 시각화 자료를 통해 본 결과와 유사함
- 대부분의 나이에서 여성 인구수가 더 많음
- 20대 이하, 40대의 경우에는 남성 인구수가 많음
- 나이대를 보면 40대이하의 인구수가 많음
- 따라서 고객의 타겟을 정할 시 30, 40대 여성을 주타겟층으로 하는 것이 가장 수요가 많을 것으로 예상됨
